```python
import numpy as np
```

```python
import networkx as nx
```

```python
weights = np.zeros(500)
weights = weights + 0.1
```

## < k > = 0.8

```python
# 0.8 = p(500 - 1) = 0.0016
G_a = nx.erdos_renyi_graph(500, 0.0016, seed=2021)
nx.draw_circular(G_a, node_size=5, width=weights)
```

```python
G_a.number_of_edges()
```

## < K > = 1

```python
# 1 = p(500 - 1) = 0.002
G_b = nx.erdos_renyi_graph(500, 0.002, seed=2021)
nx.draw_circular(G_b, node_size=5, width=weights)
```

```python
G_b.number_of_edges()
```

```python
# 8 = p(500 - 1) = 0.016
G_c = nx.erdos_renyi_graph(500, 0.016, seed=2021)
nx.draw_circular(G_c, node_size=5, width=weights)
```
